```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = pd.read_csv("../iris/IRIS.csv")
```

```python
print(data_set.head(5))
print(data_set.shape)
```

```python
data_set.drop('species' , axis = 1 , inplace = True)
```

```python
data_set.head(5)
```

```python
wcss = []
for i in range(1,8):
    model = KMeans(n_clusters = i , max_iter = 50 , init= 'k-means++')
    model.fit(data_set)
    wcss.append(model.inertia_)
    
plt.plot(wcss , 'ro-' , label = 'plot')
```

```python
### we see cluster =3 is good
model = KMeans(n_clusters = 3 , max_iter = 50 , init = 'k-means++')
y_pred = model.fit_predict(data_set)
```

```python
data_set['clusters'] = y_pred
```

```python
sns.pairplot(data_set , hue='clusters')
```

```python
### we can see that nicely clutered at n_clusteres = 3
```
